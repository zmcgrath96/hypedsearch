# Find hybrid spectra
* Load in spectrum mill results
* Look at the name of the source protein
    * If 'hybrid' in the lower case of the name, then print ID
  

```python
parent_directory = '/mnt/c/Users/zmcgr/Documents/raw_inputs'
specmill_results_filename = 'NOD2_E3_results.ssv'

results_file = os.path.join(parent_directory, specmill_results_filename)
```

```python
results_df = pd.read_csv(results_file, sep=';')
```

```python
results_df.head(5)
```

```python
name_col = 'entry_name'
```

```python
hybrid_df = results_df[results_df[name_col].str.contains('(?i)hybrid')]
```

```python
hybrid_df.head(10)
```

```python
id_col = 'filename'
hybrid_ids = hybrid_df[id_col]
```
